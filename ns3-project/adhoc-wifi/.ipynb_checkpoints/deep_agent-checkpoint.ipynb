```python
df = pd.read_csv("data.csv", index_col=0, )
df = df[df['reward']==1].reset_index()
```

```python
df.drop('index', axis=1, inplace=True)
df
```

```python
scatter_matrix(df, figsize=(15, 15));
```

```python
df[['radio', 'power', 'distance']].plot(figsize=(25, 8))
```

```python
X = df[['distance']].to_numpy()
Y = df['power'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.2,
                                                    random_state=10)

scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)  

#y_train = to_categorical(y_train, 51)
#y_test = to_categorical(y_test, 51)
```

```python
y_train.shape
#X_train.shape
```

```python
history = agent2.learn(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
plt.xlabel('Epoch')
plt.ylabel('Mean Abs Error [MPG]')
plt.plot(hist["epoch"], hist["mae"], label='Train Error')
plt.plot(hist["epoch"], hist["val_mae"], label = 'Eval Error')
plt.legend()
```

```python
plt.xlabel('Epoch')
plt.ylabel('Mean Square Error [$MPG^2$]')
plt.plot(hist["epoch"], hist["mse"], label='Train Error')
plt.plot(hist["epoch"], hist["val_mse"], label = 'Eval Error')
plt.legend()
```
