```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import argparse
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

__author__ = "Fabio Steven Tovar Ramos"
__version__ = "1.0"
__email__ = "fstovarr@unal.edu.co"
__title__ = "Data analysis"
```

# Training environment

```python
df = pd.read_csv("data/data.csv", index_col=0, )
df = df[df['reward']==1].reset_index()
df.drop('index', axis=1, inplace=True)
df.head()
```

```python
df.describe()
```

```python
df[['radio', 'power', 'distance']].plot(figsize=(25, 8))
plt.suptitle("Relation between power, radio and distance", fontsize=20)
plt.xlabel("time", fontsize=17)
```

```python
df[['power']].hist(figsize=(15, 8))
plt.suptitle("Power distribution", fontsize=20)
plt.xlabel("power", fontsize=17)
plt.ylabel("frequency", fontsize=17)
plt.savefig("plots/power_distribution_train.png")
```

```python
df[['time', 'radio']].plot(x='radio', y='time', kind='scatter', figsize=(13, 8))
plt.suptitle("Variation of radio through time", fontsize=20)
plt.xlabel("radio", fontsize=17)
plt.ylabel("time", fontsize=17)
plt.savefig("plots/time_radio_train.png")
```

# Real environment

```python
data = pd.read_csv("data/real_data.csv").reset_index().drop('index', axis=1)
data
```

```python
cd = data[['c_episode', 'c_packets', 'c_rate', 'c_power']]
cd.columns = ['nodes_movements', 'packet', 'packet_rate', 'accumulated_power']
cd.describe()
```

```python
bd = data[['b_episode', 'b_packets', 'b_rate', 'b_power']]
bd.columns = ['nodes_movements', 'packet', 'packet_rate', 'accumulated_power']
bd.describe()
```

```python
cd.describe()
```

```python
scatter_matrix(cd, figsize=(6,6))
plt.suptitle("Cognitive agent", fontsize=20)
plt.savefig("plots/scatter_cognitive_agent.png", bbox_inches='tight', dpi=100)
```

```python
bd.describe()
```

```python
fig = scatter_matrix(bd, figsize=(6, 6))
plt.suptitle("Binary agent", fontsize=20)
plt.savefig("plots/scatter_binary_agent.png", bbox_inches='tight', dpi=100)
```

```python
plt.figure(figsize=(13,8))
plt.hist([cd['packet_rate'], bd['packet_rate']], 45, rwidth=1, label=['Cognitive agent', 'Binary agent'])
plt.suptitle('Received packet rate distribution', fontsize=17)
plt.legend()
plt.savefig("plots/packet_rates.png")
```

```python
plt.figure(figsize=(10,6))
plt.plot(cd['accumulated_power'], label='Cognitive agent')
plt.plot(bd['accumulated_power'], label='Binary agent')
plt.ylabel('Accumulated power')
plt.xlabel('Nodes movements')
plt.suptitle('Power comparison', fontsize=15)
plt.legend()
plt.savefig("plots/nodes_movements_power.png")
```
