```python
def get_colors(n):
    colors = []
    for i in range(0, 360, 360 // n):
        colors.append('hsl(%d, %d%%, %d%%)' % (i, 90 + np.random.rand() * 10, 50 + np.random.rand() * 10))
    return colors
```

```python
def colouring_graph(G):
    adj = list(G.get_adjacency())
    available_colors = [True] * (len(G.vs) - 1)
    colors = [None] * (len(G.vs) - 1)
    max_color = 0
    
    for vertex in G.vs:
        for neighbor in vertex.neighbors():
            if colors[neighbor.index] != None:
                available_colors[neighbor.index] = False
    
        color = next(i for (i, value) in enumerate(available_colors) if value)
        colors[vertex.index] = color
        vertex["color"] = color
        max_color = max(max_color, color)
        
        for (n, neighbor) in enumerate(vertex.neighbors()):
            available_colors[n] = True
    
    return max_color + 1
```

```python
colors = colouring_graph(g)
for i in g.vs:
    print(i)
colors
```

```python
gen_colors = get_colors(colors)
for (i, v) in enumerate(g.vs):
    v['color'] = gen_colors[v['color']]
list(g.vs)
```
